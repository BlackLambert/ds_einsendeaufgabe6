# Data Science - Aufgabe 12 - Clustering

```python
%matplotlib inline
import numpy as mp
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
plt.rcParams['figure.figsize'] = (16, 9)

dataframe = pd.read_csv('edlich-kmeans-A0.csv')
dataframe.head(10)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(dataframe["V1"], dataframe["V2"], dataframe["V3"])
```

```python
kmeans = KMeans(n_clusters=5)
X=dataframe.values
kmeans = KMeans.fit(kmeans, X)
labels = KMeans.predict(kmeans, X)
centers = kmeans.cluster_centers_
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(dataframe["V1"], dataframe["V2"], dataframe["V3"])
ax.scatter(centers[:, 0], centers[:, 1], centers[:, 2], marker='*', c='#FF0000', s=100)
```

```python
results = pd.DataFrame([labels]).T
results
```
